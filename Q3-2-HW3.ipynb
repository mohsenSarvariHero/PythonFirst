```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of decision boundaries c
c_values = np.linspace(0, 2/3, 1000)

# Define the pdfs
def f0(x):
    return 3/2 if 0 <= x <= 2/3 else 0

def f1(x):
    return 3*x if 0 <= x <= 2/3 else 0

# Compute α and β
alpha = []
beta = []

for c in c_values:
    # Compute rejection threshold
    threshold = c / 2

    # Calculate α: Area under f0(x) for x > threshold
    alpha_value = max(0, 1 - 3/2 * threshold)  # Integral of f0(x) from threshold to 2/3
    alpha.append(alpha_value)

    # Calculate β: Area under f1(x) for x <= threshold
    beta_value = min(1, 3 * threshold * threshold / 2)  # Integral of f1(x) from 0 to threshold
    beta.append(beta_value)

# Compute Power = 1 - β
power = [1 - b for b in beta]

# Plot the power of the test as a function of decision boundary c
plt.figure(figsize=(10, 6))
plt.plot(c_values, power, label="Power of the Test")
plt.title("Power vs Decision Boundary")
plt.xlabel("Decision Boundary (c)")
plt.ylabel("Power")
plt.legend()
plt.grid()
plt.show()
```
