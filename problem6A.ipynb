```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample DataFrame (replace this with your actual `results_df` data)
# Here, we're creating some dummy data for demonstration.
# In a real scenario, you would load your own data here.
data = {
    'Quick Sort': [0.02, 0.021, 0.019, 0.018, 0.022],
    'Merge Sort': [0.03, 0.031, 0.029, 0.028, 0.032],
    'Bubble Sort': [0.12, 0.125, 0.118, 0.119, 0.123],
    'Insertion Sort': [0.08, 0.082, 0.079, 0.081, 0.083]
}
results_df = pd.DataFrame(data)

# Calculate mean, median, and standard deviation for each sorting algorithm
stats = results_df.agg(['mean', 'median', 'std']).T

# Rename the columns for better readability
stats.columns = ['Mean Duration', 'Median Duration', 'Standard Deviation']

# Sort the algorithms based on the mean duration
stats = stats.sort_values(by='Mean Duration')

# Set up the figure size for the plot
plt.figure(figsize=(12, 6))

# Create a boxplot for the duration distribution of each sorting algorithm
sns.boxplot(data=results_df)

# Adding mean and median lines for each sorting algorithm
for i, col in enumerate(results_df.columns):
    # Draw a red dashed line for the mean duration
    plt.axhline(y=stats.loc[col, 'Mean Duration'], color='r', linestyle='--',
                label=f'Mean {col}' if i == 0 else "")
    # Draw a green dotted line for the median duration
    plt.axhline(y=stats.loc[col, 'Median Duration'], color='g', linestyle=':',
                label=f'Median {col}' if i == 0 else "")

# Set the title for the plot
plt.title('Duration Distribution of Sorting Algorithms')

# Label the y-axis
plt.ylabel('Duration (seconds)')

# Label the x-axis
plt.xlabel('Sorting Algorithms')

# Set x-ticks to correspond to each algorithm name
plt.xticks(ticks=range(len(results_df.columns)), labels=results_df.columns)

# Display the legend, placing it in the upper right of the plot
plt.legend(loc='upper right')

# Add a grid to the plot for better readability
plt.grid(True)

# Show the plot
plt.show()

# Display the calculated statistics in a table format
print("Statistics Table:\n", stats)
```
