```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('dataset1csv.csv')

# Prepare the diabetes outcome column
data['Diabetes'] = data['Result of diabetes screening'].apply(lambda x: 1 if x == 'Positive' else 0)

# Boxplots
fig_age_box = px.box(data, x='Diabetes', y='Age', color='Diabetes',
                     title='Age Distribution by Diabetes Outcome',
                     labels={'Diabetes': 'Diabetes Outcome'},
                     color_discrete_sequence=['blue', 'orange'])
fig_bmi_box = px.box(data, x='Diabetes', y='BMI', color='Diabetes',
                     title='BMI Distribution by Diabetes Outcome',
                     labels={'Diabetes': 'Diabetes Outcome'},
                     color_discrete_sequence=['blue', 'orange'])

# Histograms
fig_age_hist = px.histogram(data, x='Age', color='Diabetes', barmode='overlay',
                            title='Age Distribution by Diabetes Outcome',
                            labels={'Diabetes': 'Diabetes Outcome'},
                            color_discrete_sequence=['blue', 'orange'])
fig_bmi_hist = px.histogram(data, x='BMI', color='Diabetes', barmode='overlay',
                            title='BMI Distribution by Diabetes Outcome',
                            labels={'Diabetes': 'Diabetes Outcome'},
                            color_discrete_sequence=['blue', 'orange'])

# Heatmap for correlations
correlation_data = data[['Age', 'BMI', 'Diabetes']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title('Correlation Heatmap for Age, BMI, and Diabetes')
plt.show()

# Show Plotly visuals
fig_age_box.show()
fig_bmi_box.show()
fig_age_hist.show()
fig_bmi_hist.show()
```

```python
from plotly.subplots import make_subplots
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the dataset
data = pd.read_csv('dataset1csv.csv')

# Prepare the diabetes outcome column
data['Diabetes'] = data['Result of diabetes screening'].apply(lambda x: 1 if x == 'Positive' else 0)

# Create a list to store all figures
figures = []

# Claim 1: Age and BMI as predictors
fig_age_box = px.box(data, x='Diabetes', y='Age', color='Diabetes',
                     title='Age Distribution by Diabetes Outcome',
                     labels={'Diabetes': 'Diabetes Outcome'},
                     color_discrete_sequence=['blue', 'orange'])
fig_bmi_box = px.box(data, x='Diabetes', y='BMI', color='Diabetes',
                     title='BMI Distribution by Diabetes Outcome',
                     labels={'Diabetes': 'Diabetes Outcome'},
                     color_discrete_sequence=['blue', 'orange'])
fig_age_hist = px.histogram(data, x='Age', color='Diabetes', barmode='overlay',
                            title='Age Distribution by Diabetes Outcome',
                            labels={'Diabetes': 'Diabetes Outcome'},
                            color_discrete_sequence=['blue', 'orange'])
fig_bmi_hist = px.histogram(data, x='BMI', color='Diabetes', barmode='overlay',
                            title='BMI Distribution by Diabetes Outcome',
                            labels={'Diabetes': 'Diabetes Outcome'},
                            color_discrete_sequence=['blue', 'orange'])
figures.extend([fig_age_box, fig_bmi_box, fig_age_hist, fig_bmi_hist])

# Claim 2: Class imbalance
class_counts = data['Diabetes'].value_counts()
fig_class_imbalance = px.bar(class_counts, x=class_counts.index, y=class_counts.values,
                             title='Class Distribution in Diabetes Dataset',
                             labels={'x': 'Diabetes Outcome', 'y': 'Count'},
                             text=class_counts.values)
fig_class_imbalance.update_traces(textposition='outside', marker_color=['blue', 'orange'])
fig_class_imbalance.update_layout(xaxis=dict(tickmode='array',
                                             tickvals=[0, 1],
                                             ticktext=['Negative', 'Positive']))
figures.append(fig_class_imbalance)

# Claim 3: Blood pressure predictors
fig_systolic_box = px.box(data, x='Diabetes', y='Systolic blood pressure', color='Diabetes',
                          title='Systolic Blood Pressure by Diabetes Outcome',
                          labels={'Diabetes': 'Diabetes Outcome', 'Systolic blood pressure': 'Systolic BP'},
                          color_discrete_sequence=['blue', 'orange'])
fig_diastolic_box = px.box(data, x='Diabetes', y='Diastolic blood pressure', color='Diabetes',
                           title='Diastolic Blood Pressure by Diabetes Outcome',
                           labels={'Diabetes': 'Diabetes Outcome', 'Diastolic blood pressure': 'Diastolic BP'},
                           color_discrete_sequence=['blue', 'orange'])
figures.extend([fig_systolic_box, fig_diastolic_box])

# Save all figures in one HTML file
from plotly.io import write_html
write_html(figures, file='diabetes_analysis.html', auto_open=True)
```
