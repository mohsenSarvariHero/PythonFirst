```python
import numpy as np
import matplotlib.pyplot as plt

# Define the MGF function for the normal distribution
def MGF(t, mu=0.08, sigma2=0.04):

    return np.exp(mu * t + 0.5 * sigma2 * t**2)

# Generate a range of t values for plotting
t_values = np.linspace(-1, 1, 500)
mgf_values = MGF(t_values)

# Plot the MGF
plt.figure(figsize=(8, 5))
plt.plot(t_values, mgf_values, label=r"$M_X(t) = \exp(0.08t + 0.02t^2)$", color="blue")
plt.axvline(0, color="gray", linestyle="--", alpha=0.7)
plt.title("Moment-Generating Function (MGF) of $X \sim N(0.08, 0.04)$", fontsize=14)
plt.xlabel("$t$", fontsize=12)
plt.ylabel("$M_X(t)$", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
