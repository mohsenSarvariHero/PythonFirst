```python
# Import numpy
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the null and alternative distributions
theta_null = 2
theta_alt = 2.5
acceptance_region = [0.1, 1.9]

# Define the uniform distribution PDFs
x_null = np.linspace(0, theta_null, 500)
pdf_null = np.ones_like(x_null) / theta_null  # Uniform PDF for theta = 2

x_alt = np.linspace(0, theta_alt, 500)
pdf_alt = np.ones_like(x_alt) / theta_alt  # Uniform PDF for theta = 2.5

# Plot the PDFs and highlight regions
plt.figure(figsize=(10, 6))
plt.plot(x_null, pdf_null, label='PDF under H0 (θ=2)', color='blue')
plt.plot(x_alt, pdf_alt, label='PDF under H1 (θ=2.5)', color='green', linestyle='--')
plt.fill_betweenx(pdf_null, acceptance_region[0], acceptance_region[1], color='gray', alpha=0.3, label='Acceptance Region')
plt.fill_betweenx(pdf_alt, 0, acceptance_region[0], color='red', alpha=0.4, label='Rejection Region (Left)')
plt.fill_betweenx(pdf_alt, acceptance_region[1], theta_alt, color='red', alpha=0.4, label='Rejection Region (Right)')

# Formatting the plot
plt.title('Shift in θ and Overlap Between Rejection and Acceptance Regions')
plt.xlabel('X')
plt.ylabel('Density')
plt.axvline(acceptance_region[0], color='black', linestyle='--', linewidth=1, label='Critical Value 0.1')
plt.axvline(acceptance_region[1], color='black', linestyle='--', linewidth=1, label='Critical Value 1.9')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
