```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parameters for the Gamma distribution
shape_param = 5  # k
scale_param = 200  # θ
num_observations = 100

# (a) Simulate data from a Gamma distribution
simulated_data = np.random.gamma(shape=shape_param, scale=scale_param, size=num_observations)

# (b) Compute sample quantiles
sorted_data = np.sort(simulated_data)
probabilities = np.linspace(0.01, 0.99, num_observations)

# Compute theoretical quantiles using the Gamma distribution's inverse CDF
theoretical_quantiles = stats.gamma.ppf(probabilities, a=shape_param, scale=scale_param)

# (c) Generate Q-Q plot comparing sample data against theoretical quantiles
plt.figure(figsize=(8, 6))
plt.plot(theoretical_quantiles, sorted_data, 'o', label='Sample Data')
plt.plot([min(theoretical_quantiles), max(theoretical_quantiles)],
         [min(theoretical_quantiles), max(theoretical_quantiles)], 'r--', label='Theoretical Line')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.title('Q-Q Plot for Gamma Distribution')
plt.legend()
plt.grid()
plt.show()

# (d) Perform Box-Cox transformation on the data
transformed_data, lambda_bc = stats.boxcox(simulated_data[simulated_data > 0])  # Box-Cox requires positive values

# Compute sample quantiles for transformed data
sorted_transformed_data = np.sort(transformed_data)

# Generate Q-Q plot for transformed data
plt.figure(figsize=(8, 6))
plt.plot(theoretical_quantiles, sorted_transformed_data, 'o', label='Transformed Data')
plt.plot([min(theoretical_quantiles), max(theoretical_quantiles)],
         [min(theoretical_quantiles), max(theoretical_quantiles)], 'r--', label='Theoretical Line')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Transformed Quantiles')
plt.title(f'Q-Q Plot for Box-Cox Transformed Data (Lambda = {lambda_bc:.2f})')
plt.legend()
plt.grid()
plt.show()
```
