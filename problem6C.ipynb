```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample DataFrame (replace this with your actual `results_df` data)
data = {
    'Quick Sort': [0.02, 0.021, 0.019, 0.018, 0.022],
    'Merge Sort': [0.03, 0.031, 0.029, 0.028, 0.032],
    'Bubble Sort': [0.12, 0.125, 0.118, 0.119, 0.123],
    'Insertion Sort': [0.08, 0.082, 0.079, 0.081, 0.083]
}
results_df = pd.DataFrame(data)

# Plot the boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=results_df)
plt.title("Boxplots of Sorting Algorithm Durations")
plt.ylabel("Duration (seconds)")
plt.xlabel("Sorting Algorithms")
plt.show()
```

```python
# Count the outliers for each algorithm
outlier_counts = {}

for col in results_df.columns:
    Q1 = results_df[col].quantile(0.25)
    Q3 = results_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = results_df[(results_df[col] < lower_bound) | (results_df[col] > upper_bound)][col]
    outlier_counts[col] = len(outliers)

# Display the count of outliers for each algorithm
print("Outlier counts:\n", outlier_counts)

# Identify the algorithm with the most outliers
most_outliers_algorithm = max(outlier_counts, key=outlier_counts.get)
most_outliers_count = outlier_counts[most_outliers_algorithm]

print(f"\nAlgorithm with the most outliers: {most_outliers_algorithm} ({most_outliers_count} outliers)")
```
