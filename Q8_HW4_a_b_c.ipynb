```python
import numpy as np
import matplotlib.pyplot as plt

def compute_ecdf(data):
    """
    Compute the empirical cumulative distribution function (eCDF).

    Parameters:
        data (array-like): The dataset to compute the eCDF for.

    Returns:
        x (np.ndarray): Sorted data points.
        y (np.ndarray): Corresponding eCDF values.
    """
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n + 1) / n
    return x, y

def compute_confidence_band(n, alpha=0.05):
    """
    Compute the margin of error (epsilon) and confidence band for the eCDF.

    Parameters:
        n (int): Sample size.
        alpha (float): Significance level (default: 0.05 for 95% confidence).

    Returns:
        epsilon (float): Margin of error for the confidence band.
    """
    epsilon = np.sqrt(np.log(2 / alpha) / (2 * n))
    return epsilon

# Dataset
data = [120, 135, 140, 125, 130, 150, 145, 155, 160, 125, 135, 140, 145, 150, 155, 160, 165, 170, 175, 130,
        135, 140, 145, 150, 125, 130, 140, 135, 145, 150, 155, 165, 140, 145, 150, 135, 140, 145, 150, 155,
        140, 130, 120, 150, 155, 145, 135, 125, 130, 140]

# Compute eCDF
x, y = compute_ecdf(data)

# Compute Confidence Band
n = len(data)
alpha = 0.05
epsilon = compute_confidence_band(n, alpha)

# Compute Upper and Lower Bounds
lower_bound = np.maximum(0, y - epsilon)
upper_bound = np.minimum(1, y + epsilon)

# Plot eCDF and Confidence Band
plt.figure(figsize=(10, 6))
plt.step(x, y, where='post', label='eCDF', color='blue')
plt.step(x, lower_bound, where='post', linestyle='--', color='red', label='Lower 95% CI')
plt.step(x, upper_bound, where='post', linestyle='--', color='green', label='Upper 95% CI')
plt.xlabel('Data')
plt.ylabel('eCDF')
plt.title('Empirical Cumulative Distribution Function with 95% Confidence Band')
plt.grid()
plt.legend()
plt.show()

# Print results
print(f"Margin of error (epsilon): {epsilon:.4f}")
```
