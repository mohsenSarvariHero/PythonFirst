```python
import numpy as np
import matplotlib.pyplot as plt
# Parameters
mean_pressure = 75348 # Mean pressure in Pascals
std_dev = 25 # Standard deviation in Pascals
n = 100 # Number of measurements
# Simulate the pressure measurements
np.random.seed(0) # For reproducibility
pressure_measurements = np.random.normal(mean_pressure, std_dev, n)
# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(pressure_measurements, bins=15, color='skyblue',
edgecolor='black', alpha=0.7)
plt.axvline(mean_pressure, color='red', linestyle='dashed', linewidth=1.5,
label='Mean (75,348 Pascals)')
plt.axvline(mean_pressure + std_dev, color='green', linestyle='dashed',
linewidth=1.5, label='Mean + SD (75,373 Pascals)')
plt.axvline(mean_pressure - std_dev, color='green', linestyle='dashed',
linewidth=1.5, label='Mean - SD (75,323 Pascals)')
plt.title('Histogram of Pressure Measurements')
plt.xlabel('Pressure (Pascals)')
plt.ylabel('Frequency')
plt.legend()
plt.grid()
plt.show()
```
