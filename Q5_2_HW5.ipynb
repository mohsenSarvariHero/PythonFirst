```python
# Re-importing necessary library for plotting after code reset
import matplotlib.pyplot as plt

# Data: P-values and calculated thresholds
p_values = [0.005, 0.009, 0.019, 0.022, 0.051, 0.101, 0.361, 0.387]
thresholds = [(r + 1) / 8 * 0.05 for r in range(len(p_values))]  # BH thresholds

# Plot p-values and thresholds
ranks = range(1, len(p_values) + 1)
plt.figure(figsize=(8, 6))

plt.plot(ranks, p_values, marker='o', label='P-values', linestyle='--', color='blue')
plt.plot(ranks, thresholds, marker='x', label='BH Thresholds', linestyle='-', color='red')

# Highlight significant p-values (those below the threshold)
significant_indices = [i for i, p in enumerate(p_values) if p <= thresholds[i]]
plt.scatter(
    [ranks[i] for i in significant_indices],
    [p_values[i] for i in significant_indices],
    color='green',
    label='Significant P-values',
    zorder=5,
)

# Add labels, title, and legend
plt.axhline(0.05, color='gray', linestyle=':', label='Significance Level (α = 0.05)')
plt.xlabel("Rank")
plt.ylabel("P-value")
plt.title("P-values and Benjamini-Hochberg Thresholds")
plt.legend()
plt.grid(True)
plt.xticks(ranks)
plt.ylim(0, 0.1)

# Show the plot
plt.show()
```
