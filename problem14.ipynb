```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
num_genes = 10
num_simulations = 100

# Increasing robustness: Probability of change decreases from gene 1 to gene 10
# Assuming robustness increases linearly, we could scale probabilities from 0.9 to 0.1
change_probabilities = np.linspace(0.9, 0.1, num_genes)

# Simulate changes for each gene 100 times
change_counts = np.zeros(num_genes)
for _ in range(num_simulations):
    # Sample a change for each gene based on its robustness probability
    changes = np.random.uniform(0, 1, num_genes) < change_probabilities
    change_counts += changes

# Normalize by the number of simulations to get frequencies
change_frequencies = change_counts / num_simulations

# Plot the distribution of change frequency for gene 5
gene_5_changes = [1 if np.random.uniform(0, 1) < change_probabilities[4] else 0 for _ in range(num_simulations)]
sns.histplot(gene_5_changes, bins=10, kde=True)
plt.title("Distribution of Change Frequency for Gene 5")
plt.xlabel("Change Occurrence")
plt.ylabel("Frequency")
plt.show()

# Analyze and describe the distribution
print(f"Average change frequency for Gene 5: {np.mean(gene_5_changes)}")
print(f"Variance of change frequency for Gene 5: {np.var(gene_5_changes)}")
```
