```python
import numpy as np
import matplotlib.pyplot as plt

# Define simulation parameters
num_samples = 100000  # Number of samples for Monte Carlo simulation

# Simulate enemy movement
x_enemy = np.random.uniform(0, 1, num_samples)  # Enemy's movement along x-axis
y_enemy = np.random.uniform(0, 1 - x_enemy)    # Enemy's movement along y-axis

# Farming along x-axis
def farming_x_area(x):
    """Calculates the farming area along x-axis."""
    return 1 - x  # Area of a vertical strip from (x, 0) to (x, 1-x)

successful_area_x = np.mean(farming_x_area(x_enemy))

# Farming along y-axis
def farming_y_area(y):
    """Calculates the farming area along y-axis."""
    return 1 - y  # Area of a horizontal strip from (0, y) to (1-y, y)

successful_area_y = np.mean(farming_y_area(y_enemy))

# Results
print(f"Expected successful farming area along x-axis: {successful_area_x:.3f}")
print(f"Expected successful farming area along y-axis: {successful_area_y:.3f}")

# Plotting the triangular farm and enemy movements
plt.figure(figsize=(8, 8))
plt.scatter(x_enemy[:1000], y_enemy[:1000], s=1, alpha=0.5, label="Enemy movement")
plt.plot([0, 1, 0], [0, 0, 1], color="black", label="Farm boundary (X + Y = 1)")
plt.title("Enemy Movement in the Farm")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
