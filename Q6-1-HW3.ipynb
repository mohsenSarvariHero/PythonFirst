```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the uniform distribution
theta = 2
rejection_region = [0.1, 1.9]

# Define the uniform distribution PDF
x = np.linspace(0, theta, 500)
pdf = np.ones_like(x) / theta  # Uniform PDF over [0, theta]

# Compute the probability of a Type I error
alpha_low = rejection_region[0] / theta  # Area under PDF from [0, 0.1]
alpha_high = (theta - rejection_region[1]) / theta  # Area under PDF from [1.9, 2]
alpha = alpha_low + alpha_high

# Plot the uniform PDF and rejection regions
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label='PDF of U[0, θ]', color='blue')
plt.fill_betweenx(pdf, 0, rejection_region[0], color='red', alpha=0.4, label='Rejection Region [0, 0.1]')
plt.fill_betweenx(pdf, rejection_region[1], theta, color='red', alpha=0.4, label='Rejection Region [1.9, 2]')

# Formatting the plot
plt.title('Uniform PDF with Rejection Regions for θ=2')
plt.xlabel('X')
plt.ylabel('Density')
plt.axvline(0.1, color='black', linestyle='--', linewidth=1, label='Critical Value 0.1')
plt.axvline(1.9, color='black', linestyle='--', linewidth=1, label='Critical Value 1.9')
plt.legend()
plt.grid(alpha=0.3)
plt.show()


```
